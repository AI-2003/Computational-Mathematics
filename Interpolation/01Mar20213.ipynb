Interpolaci√≥n a trozos (cubic splines)

```python
from scipy.interpolate import make_interp_spline as makeSpline
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda t: 1/(1+t**2)
I = [-3,3]

ts = np.linspace(I[0], I[1], 7)
tsFino = np.linspace(I[0], I[1], 129)
xs = f(ts)

S = makeSpline(ts, f(ts), bc_type=([(1,0.06)],[(1,-0.06)]))




plt.plot(tsFino, f(tsFino), label="original")
plt.plot(tsFino, S(tsFino), label="new")
plt.legend()
```
