```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Evaluate Taylor polinomials for f(x)=e**x
def evaTm(xs, m):
    a = 0.0
    sumando = np.exp(a)*np.ones(xs.shape)
    res = sumando
    for k in range(1,m+1):
        sumando = sumando * (xs-a)/k
        res += sumando
    return res
```

```python
# Evaluation
I = [-4,2]
xs = np.linspace(I[0], I[1], 129)
plt.plot(xs, np.exp(xs), '--', label='exp')
for i in range(6):
    plt.plot(xs, evaTm(xs,i), label='m='+str(i))
plt.legend()
```
