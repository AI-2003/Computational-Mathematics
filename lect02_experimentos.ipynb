# lect02: Zoom y redondeo en precision simple (single)

```python
# includes
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0.988, 1.012, 1001)
coeff = np.array([1.0, -7.0, 21., -35., 35., -21., 7., -1.])
#valores = coeff[0]*x**7 + coeff[1]*x**6 + coeff[2]*x**5 + coeff[3]*x**4 + coeff[4]*x**3 + coeff[5]*x**2 + coeff[6]*x + coeff[7]
valores = np.zeros(x.shape)
for i in range(len(coeff)):
    valores += coeff[i]*x**(7-i)
#print(valores)

f, ax = plt.subplots()
f.set_figheight(9)
f.set_figwidth(16)
plt.grid()
plt.plot(x, valores, '.')
plt.plot(x, (x-1.)**7, linewidth=5)
plt.legend(['expanded', 'not expanded'], fontsize=20)
plt.title('(x-1)^7 expanded vs. not expanded')
```

## Redondeo en "single precision"

```python
import numpy as np
single = lambda r:  np.float32(r)    # redondear numero a 7 cifras decimales.
```

```python
# suma: a + b  en  R7
a = 5037654.0
b = 25.874

(a+b-single(a + b))/(a+b)
```

```python
# resta: a - b (cifras se pierden cuando los numero se almacenan)
a = 5037679.874
b = 5037679.0

resta_double = a - b
print('resta DP = ', resta_double)
resta_single = single(a) - single(b)
print('resta SP = ', resta_single) 
error_abs = abs(resta_double - resta_single)
error_rel = error_abs/abs(resta_double)
print('absolute error = ', error_abs)
print('relative error = ', error_rel)
```

```python
# resta a + b - c (cifras importantes se pierden por redondear la primera suma)
a = 5037654.0
b = 25.874
c = 5037679.0
print('resultado exacto: ', 0.874)
print('resultado R16   : ', a + b - c)
print('resultado R7    : ', single(a) + single(b) - single(c))
```
